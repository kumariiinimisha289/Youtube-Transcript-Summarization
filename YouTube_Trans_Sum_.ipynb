```python
from transformers import pipeline
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
youtube_video = "https://www.youtube.com/watch?v=qfaficghpom&list=plc3y8-rfhvwgg3vayjghgnmodb54rxok3&ab_channel=codevolution"
```

```python
video_id = youtube_video.split('v=')[1].split('&')[0]
```

```python
video_id
```

```python
from IPython.display import YouTubeVideo
YouTubeVideo(video_id)
```

```python
from youtube_transcript_api import get_transcript
transcript = get_transcript(video_id)
```

```python
# transcript[0:5]
```

```python
result = ""
for i in transcript:
  result += ' ' + i['text']

# print(result)
print(len(result))
```

Model

```python
summarizer = pipeline('summarization', model="philschmid/bart-large-cnn-samsum")
```

```python
num_iters = int(len(result)/1000)
print(num_iters)
summarized_text = []
for i in range(0, num_iters+1):
  start = 0
  start = i * 1000
  end = (i+1)*1000
  # print("Input text: " + result[start:end])
  out = summarizer(result[start:end], max_length=121)
  # print("Out Array: ", out)
  out = out[0]
  out = out['summary_text']
  # print("Summarized text: "+out)
  summarized_text.append(out)

# print(summarized_text)
```

```python
print("Original Length:", len(result))
print("Length of Summarized Text:", len(str(summarized_text)))
```
